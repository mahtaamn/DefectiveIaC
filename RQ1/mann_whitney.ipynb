```python
from scipy.stats import mannwhitneyu
import numpy as np
import pandas as pd


opens = pd.read_csv("IF_Data/IST_OST.csv")
wiki = pd.read_csv("IF_Data/IST_WIK.csv")
mirantis = pd.read_csv("IF_Data/IST_MIR.csv")
mozilla = pd.read_csv("IF_Data/IST_MOZ.csv")


ds = [mirantis, mozilla, opens, wiki]
columns = ['org', 'file_', 'URL', 'File', 'Lines_of_code', 'Require', 'Ensure',
               'Include', 'Attribute', 'Hard_coded_string', 'Comment', 'Command',
               'File_mode', 'SSH_KEY', 'defect_status']
for df in ds:
        org_dict = {}
        defective = df[df["defect_status"]==1]
        neutral = df[df["defect_status"]==0]

        for feature in columns[2:14]:
            feat_dict = {}
            fdf_d = defective[feature]
            fdf_nd = neutral[feature]
            U1, p = mannwhitneyu(fdf_d,fdf_nd)
            feat_dict["p"] = p
            feat_dict["p_round"] = round(p,4)
            feat_dict["U1"] = U1
            org_dict[feature] = feat_dict

        out_df = pd.DataFrame.from_dict(org_dict).T
        print("file::::",df.iloc[0]["org"])
        print(out_df)
        #out_df.to_csv(df.iloc[0]["org"]+"_MW.csv")
```
